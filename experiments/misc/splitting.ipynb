```python
from DataGenerator import DataGenerator
import numpy as np
import matplotlib.pyplot as plt

gen = DataGenerator(control=True,huntington=True)
names = []
for s in gen.names:
    names += s
cnts = [len(gen.getDatapoint(n)[0]) for n in names]

bins, edges = np.histogram(cnts,10)

fig, p = plt.subplots(1,1)
fig.set_size_inches(8, 7)
p.stairs(bins,edges,fill=True)
plt.show()
plt.close()
```
